## 1. Создание и изменение массивов

Создайте массив NumPy размером 3×3, содержащий числа от 1 до 9. Умножьте все элементы массива на 2, а затем замените все элементы больше 10 на 0. Выведите итоговый массив.

```python
import numpy as np
arr = np.arange(1, 10).reshape(3, 3)
print("Исходный массив:\n", arr)
arr *= 2
arr[arr > 10] = 0
print("\nИтоговый массив:\n", arr)
```

## 4. Генерация и работа с линеными последовательностями

Создайте массив из 50 чисел, равномерно распределенных от -10 до 10. Вычислите сумму всех элементов, сумму положительных элементов и сумму отрицательных элементов. Выведите результаты.

```python
import numpy as np
arr = np.linspace(-10, 10, 50)
print("Массив:", arr)
total_sum = np.sum(arr)
pos_sum = np.sum(arr[arr > 0])
neg_sum = np.sum(arr[arr < 0])
print("\nОбщая сумма всех элементов:", total_sum)
print("Сумма положительных элементов:", pos_sum)
print("Сумма отрицательных элементов:", neg_sum)
```

## 10. Системы линейных уравнений

Решить систему линейных уравнений.

\begin{cases}
2x + 3y - z = 5 \\
4x - y + 2z = 6 \\
-3x + 5y + 4z = -2
\end{cases}

Используйте матричное представление Ax = B , где A – матрица коэффициентов, x – вектор неизвестных, B – вектор правой части. Решите систему с помощью np.linalg.solve и выведите результат.

```python
import numpy as np
A = np.array([
    [2, 3, -1],
    [4, -1, 2],
    [-3, 5, 4]
])
B = np.array([5, 6, -2])
x = np.linalg.solve(A, B)
print("Решение системы уравнений:\n", x)
```

## 11. Индивидуальное задание

Каждое задание предусматривает построение системы линейных уравнений. Решите полученную систему уравнений с использованием библиотеки NumPy. Для решения системы используйте метод Крамера и матричный метод. Сравните полученные результаты, с результатами, полученными с помощью np.linalg.solve.

**№9. Отпимальный выбор топлива.** Автомобиль использует три вида топлива в разных пропорциях. Бензин даёт 30 МДж энергии на литр, дизель — 35 МДж, газ — 25 МДж. Для поездки требуется 100 МДж энергии. Водитель хочет использовать бензина в два раза больше, чем газа, а количество дизеля равно количеству газа. Сколько литров каждого топлива нужно залить?

Решение:
Пусть:

1. x - литры бензина
2. y - литры дизеля
3. z - литры газа

Система уравнений:

1. Энергия: 30x+35y+25x=100
2. Бензин/газ: x=2z или x-2z=0
3. Дизель/газ: y=z или y-z=0

```python
import numpy as np
A = np.array([
    [30, 35, 25],
    [1, 0, -2],
    [0, 1, -1]
])
B = np.array([100, 0, 0])
print("Матрица коэффициентов A:")
print(A)
print("\nВектор правой части B:")
print(B)
```

1. Метод Крамера

```python
det_A = np.linalg.det(A)
A_x = A.copy()
A_x[:, 0] = B

A_y = A.copy()
A_y[:, 1] = B

A_z = A.copy()
A_z[:, 2] = B

det_Ax = np.linalg.det(A_x)
det_Ay = np.linalg.det(A_y) 
det_Az = np.linalg.det(A_z)

x_kramer = det_Ax / det_A
y_kramer = det_Ay / det_A
z_kramer = det_Az / det_A

print(f"Бензин: {x_kramer:.2f} литров")
print(f"Дизель: {y_kramer:.2f} литров")
print(f"Газ: {z_kramer:.2f} литров")
```

2. Матричный метод (A⁻¹B)

```python
A_inv = np.linalg.inv(A)
sol_matrix = A_inv @ B
    
print(f"Бензин: {sol_matrix[0]:.2f} литров")
print(f"Дизель: {sol_matrix[1]:.2f} литров")
print(f"Газ: {sol_matrix[2]:.2f} литров")
```

3. Метод np.linalg.solve

```python
sol_solve = np.linalg.solve(A, B)

print(f"Бензин: {sol_solve[0]:.2f} литров")
print(f"Дизель: {sol_solve[1]:.2f} литров") 
print(f"Газ: {sol_solve[2]:.2f} литров")
```

## Все методы дали одинаковый результат:
Бензин: 1.67 литров

Дизель: 0.83 литров
